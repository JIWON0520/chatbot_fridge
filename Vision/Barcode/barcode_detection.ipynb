```python
import cv2
import os, io
from matplotlib import pyplot as plt
import pyzbar.pyzbar as pyzbar
```

```python
# imshow function (cv2 image -> plt image -> plt.show)
# cv2 image를 plt image로 출력 
def imshow(in_img):
    out_img = cv2.cvtColor(in_img,cv2.COLOR_BGR2RGB) 
    plt.figure(figsize=(20,20))
    plt.imshow(out_img)
    plt.axis('off')
    plt.show()
```

```python
def detect_barcode(image):
  if image.shape[0]>=1000:
    image=cv2.resize(image,dsize=(0,0),fx=0.2,fy=0.2)

  barcodeCnt=len(pyzbar.decode(image))

  if barcodeCnt==0: 
    return 0

  barcodeList=pyzbar.decode(image)

  maxArea=0
  for barcode in barcodeList:
    (_,_,w,h)=barcode.rect
    barcodeArea=w*h
    if barcodeArea>maxArea:
      maxArea=barcodeArea
      result=barcode.data.decode('utf-8')
  
  return result
```

```python
# 파일 이름 및 경로 설정
FILE_NAME='1.jpg' 
FOLDER_PATH="pictures2" 
FILE_PATH=os.path.join(FOLDER_PATH, FILE_NAME)
```

```python
# 이미지 객체 생성 및 확인 
img=cv2.imread(FILE_PATH)
imshow(img)
```

```python
rst=detect_barcode(img)
print(rst)
print(len(rst))
```
